Example on how to import the generated dataset.

This shows how to import the global labels array I created. 

I will then show  how to import the images and place them in one array; how to get the label of the images and the square name too.

```python
import numpy as np
import os
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

filepath = os.getcwd()+"/Data Generation/Data Generated/Dataset PreProcessed/generatedData.npy"
labels = np.load(filepath, allow_pickle=True)

print(labels[0:5])
```

Here is how to import all images in one array.

```python
# Get all images in one array
images = labels[1:, 1]
```

Here is how to vizualize one image and get its label and square name (in one hot encoded)

```python
# Trainign exmaple number
training_example = 0

# Square number of the trainign example to visialize. A1 = 0; H8 = 63
square_number = 0

# Get the index of the image in the array
index = training_example*64 + square_number

print(images[index].shape)

print("Label of this piece is:", labels[index+1][2])

print("Square of this piece is:", labels[index+1][3])

# Assuming images array is already defined
plt.imshow(images[index])
plt.show()
```
